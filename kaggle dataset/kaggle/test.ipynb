```python
count=0;
for i in data:
    if data[i].isnull().sum()>1:
        count+=1
        print(i +  '----'+ str(data[i].isnull().sum()))
print(count)        
```

```python
for i in data:
    if data[i].isnull().sum()>1 and str(data[i].dtype)=='object':
        data[i]=data[i].fillna(data[i].mode()[0])
```

```python
count=0;
for i in data:
    if data[i].isnull().sum()>1:
        count+=1
        print(i +  '----'+ str(data[i].isnull().sum()))
print(count) 
```

```python
data['LotFrontage']=data['LotFrontage'].fillna(data['LotFrontage'].mean())
```

```python
count=0;
for i in data:
    if data[i].isnull().sum()>1:
        count+=1
        print(i +  '----'+ str(data[i].isnull().sum()))
print(count) 
```
